```python
import matplotlib.pyplot as plt
import torch
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from util import Trainer

%matplotlib inline
```

```python
data = load_iris()
X = data['data']
Y = data['target']
```

# Exc 1

# Model 1

## 1.1 Define model

```python
network1 =  torch.nn.Sequential(
               torch.nn.Linear(4,1),
           )
```

## 1.2 Train model

```python
t1 = Trainer(model=network1)

t1.fit(X,Y)
```

## 1.3 Plot result

```python
t1.plot_loss()
```

# Model 2

## 2.1 Define model

```python
network2 = torch.nn.Sequential(
    torch.nn.Linear(4,10),
    torch.nn.Linear(10,1)
)
```

## 2.2 Train model

```python
t2 = Trainer(model=network2)

t2.fit(X,Y)
```

## 2.3 Plot result

# Model 3

## 3.1 Define model

```python
network3 = torch.nn.Sequential(
    torch.nn.Linear(4,10),
    torch.nn.Linear(10,10),
    torch.nn.Linear(10,1)
)
```

## 3.2 Train model

```python
t3 = Trainer(model=network3)
t3.fit(X,Y)
```

## 3.3 Plot result

## Plot compare

```python
plt.plot(t1.loss_history,label="t1")
plt.plot(t2.loss_history,label="t2_1")
plt.plot(t3.loss_history,label="t3")
plt.legend()
```

# Exc 3

```python
Network2_2 = torch.nn.Sequential(
    torch.nn.Linear(4,10),
    torch.nn.Sigmoid(),
    torch.nn.Linear(10,1)
)
```

## Train model

```python
t2_2 = Trainer(model=Network2_2)
t2_2.optimizer = torch.optim.SGD(Network2_2.parameters(), lr=0.01)

t2_2.fit(X,Y)
```

## Plot result

```python
t2_2.plot_loss()
```

# Exc 3.2
## Other Optimizer (Solver)

# Nesterov

```python
t_Nesterov = Trainer(model=Network2_2)
t_Nesterov.optimizer = torch.optim.SGD(Network2_2.parameters(), lr=0.01,nesterov=True, momentum=.01)

t_Nesterov.fit(X,Y)
```

# Momentum

```python
t_Momentum = Trainer(model=Network2_2)
t_Momentum.optimizer = torch.optim.SGD(Network2_2.parameters(), lr=0.01, momentum=.01)

t_Momentum.fit(X,Y)
```

# RMSProp

```python
t_RMSProp = Trainer(model=Network2_2)
t_RMSProp.optimizer = torch.optim.RMSprop(Network2_2.parameters(), lr=0.01)

t_RMSProp.fit(X,Y)
```

# Adam

```python
t_Adam = Trainer(model=Network2_2)
t_Adam.optimizer = torch.optim.Adam(Network2_2.parameters(), lr=0.01)

t_Adam.fit(X,Y)
```

# Adagrad

```python
t_Adagrad = Trainer(model=Network2_2)
t_Adagrad.optimizer = torch.optim.Adagrad(Network2_2.parameters(), lr=0.01)

t_Adagrad.fit(X,Y)
```
